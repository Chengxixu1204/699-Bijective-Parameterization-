```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from numpy import arccos, array, dot, pi, cross
from numpy.linalg import norm
from scipy.spatial import ConvexHull
import numpy as np

def normalize(v):
    norm = np.linalg.norm(v)
    if norm == 0: 
       return v
    return v / norm

def create_icosphere(subdivisions, radius=1):
    t = (1.0 + 5.0 ** 0.5) / 2.0

    # Vertices of a base icosahedron
    vertices = [
        [-1,  t,  0],
        [ 1,  t,  0],
        [-1, -t,  0],
        [ 1, -t,  0],
        [ 0, -1,  t],
        [ 0,  1,  t],
        [ 0, -1, -t],
        [ 0,  1, -t],
        [ t,  0, -1],
        [ t,  0,  1],
        [-t,  0, -1],
        [-t,  0,  1],
    ]
    vertices = [vertex/norm(vertex) * radius for vertex in vertices]

    # Faces of the base icosahedron
    faces = [
        [0, 11, 5],
        [0, 5, 1],
        [0, 1, 7],
        [0, 7, 10],
        [0, 10, 11],
        [1, 5, 9],
        [5, 11, 4],
        [11, 10, 2],
        [10, 7, 6],
        [7, 1, 8],
        [3, 9, 4],
        [3, 4, 2],
        [3, 2, 6],
        [3, 6, 8],
        [3, 8, 9],
        [4, 9, 5],
        [2, 4, 11],
        [6, 2, 10],
        [8, 6, 7],
        [9, 8, 1],
    ]

    # Subdivide faces
    for _ in range(subdivisions):
        faces_subdiv = []
        new_vertices = {}
        for tri in faces:
            # Create midpoints and normalize them (to lie on the sphere)
            mid = [
                normalize((vertices[tri[i]] + vertices[tri[j]]) / 2.0) * radius
                for i, j in [(0, 1), (1, 2), (2, 0)]
            ]
            # Get or create index for each midpoint
            mid_idx = []
            for point in mid:
                point_tuple = tuple(point)
                if point_tuple not in new_vertices:
                    new_vertices[point_tuple] = len(vertices)
                    vertices.append(point)
                mid_idx.append(new_vertices[point_tuple])

            # Create new faces
            faces_subdiv.extend([
                [tri[0], mid_idx[0], mid_idx[2]],
                [tri[1], mid_idx[0], mid_idx[1]],
                [tri[2], mid_idx[1], mid_idx[2]],
                [mid_idx[0], mid_idx[1], mid_idx[2]]
            ])
        faces = faces_subdiv

    return array(vertices), array(faces)

# Create icosphere with 2 subdivisions
vertices, faces = create_icosphere(2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a Poly3DCollection object with the faces of the icosphere
poly3d = Poly3DCollection(vertices[faces], linewidths=1, edgecolors='r', alpha=.25)

# Add the Poly3DCollection object to the 3D plot
ax.add_collection3d(poly3d)

# Plot the vertices of the icosphere
ax.scatter(vertices[:,0], vertices[:,1], vertices[:,2], marker='o')

# Display the plot
plt.show()

# Save the icosphere to an OBJ file
with open('icosphere.obj', 'w') as f:
    for vertex in vertices:
        f.write(f'v {" ".join(map(str, vertex))}\n')
    for face in faces:
        f.write(f'f {" ".join(map(str, [i + 1 for i in face]))}\n')
        
print("OBJ file created successfully!")
```

```python
import numpy as np


def compute_triangle_area(triangle_vertices):
    a, b, c = triangle_vertices
    return 0.5 * np.linalg.norm(np.cross(b - a, c - a))

def compute_singular_values(P, U):
    # Create matrices to solve for the affine transformation matrix
    P_matrix = np.array([
        [P[0,0], P[1,0], 0, 0, 1, 0], 
        [0, 0, P[0,0], P[1,0], 0, 1],
        [P[0,1], P[1,1], 0, 0, 1, 0], 
        [0, 0, P[0,1], P[1,1], 0, 1],
        [P[0,2], P[1,2], 0, 0, 1, 0], 
        [0, 0, P[0,2], P[1,2], 0, 1],
    ])
    
    U_vector = U.T.flatten()
    
    # Solve for the transformation matrix elements
    phi_elements, _, _, _ = np.linalg.lstsq(P_matrix, U_vector, rcond=None)
    
    # Reconstruct the phi matrix
    phi = np.array([
        [phi_elements[0], phi_elements[1], phi_elements[4]],
        [phi_elements[2], phi_elements[3], phi_elements[5]]
    ])

    # Perform singular value decomposition to find the singular values
    U_mat, s, Vt = np.linalg.svd(phi[:2, :2])

    return s

# Naive initial 2D parameterization: using x, y coordinates of the 3D vertices
parameterized_vertices_2d = vertices[:, :2]
print(parameterized_vertices_2d)

triangle_3d = np.transpose(vertices[faces[0]])
triangle_2d = np.transpose(parameterized_vertices_2d[faces[0]])

# Calculate the singular values for this triangle
sigma1, sigma2 = compute_singular_values(triangle_3d, triangle_2d)
print("Singular values: 1 =", sigma1, ", 2 =", sigma2)
```

```python
def compute_distortion_metrics(triangle_3d, triangle_2d):
    # Extracting vertices for easier reference
    P1, P2, P3 = np.array(triangle_3d)
    U1, U2, U3 = np.array(triangle_2d)

    # Computing areas of the triangles
    area_3d = compute_triangle_area(triangle_3d)
    area_2d = compute_triangle_area(triangle_2d)
    

    # Computing lengths of edges in both 3D and 2D spaces
    l_P1P2 = np.linalg.norm(P2 - P1)
    l_P1P3 = np.linalg.norm(P3 - P1)
    l_U1U2 = np.linalg.norm(U2 - U1)
    l_U1U3 = np.linalg.norm(U3 - U1)

    # Computing dot products of the edge vectors
    dot_U = np.dot(U3 - U1, U2 - U1)
    dot_P = np.dot(P3 - P1, P2 - P1)

    # Computing the term1 and term2 as per the formulas
    term1 = (l_U1U3**2 * l_P1P2**2 + l_U1U2**2 * l_P1P3**2) / (4 * area_3d**2) - (dot_U * dot_P)**2 / (2 * area_3d**2)
    distortion_metric = (1 + area_3d**2 / area_2d**2) * ((l_U1U3**2 * l_P1P2**2 + l_U1U2**2 * l_P1P3**2) / (4 * area_3d) - (dot_U * dot_P)**2 / (2 * area_3d))
    
    # Getting sigma values using your existing method
    sigma1, sigma2 = compute_singular_values(triangle_3d, triangle_2d)

    # Computing existing metrics
    conformal_distortion = max(sigma1/sigma2, sigma2/sigma1)
    authalic_distortion = area_2d / area_3d
    
    U2_U3 = U2 - U3
    U2_U3_perp = np.array([-U2_U3[1], U2_U3[0]])

    # Finding the partial derivative of termA with respect to U1
    delta_P_squared = area_3d**2
    delta_U_squared = area_2d**2
    d_termA_d_U1 = -delta_P_squared / (delta_U_squared ** 1.5) * U2_U3_perp
    
    angle_P1 = np.arccos(np.dot(P2 - P1, P3 - P1) / (np.linalg.norm(P2 - P1) * np.linalg.norm(P3 - P1)))
    angle_P2 = np.arccos(np.dot(P1 - P2, P3 - P2) / (np.linalg.norm(P1 - P2) * np.linalg.norm(P3 - P2)))
    angle_P3 = np.arccos(np.dot(P1 - P3, P2 - P3) / (np.linalg.norm(P1 - P3) * np.linalg.norm(P2 - P3)))

    # Finding the partial derivative of termB with respect to U1 using cotangent weights
    cot_theta2 = 1 / np.tan(angle_P2)
    cot_theta3 = 1 / np.tan(angle_P3)
    d_termB_d_U1 = -cot_theta2 * U3 - cot_theta3 * U2 + (cot_theta2 + cot_theta3) * U1

    
    # Finding termA and termB from your distortion metric equation
    termA = 1 + delta_P_squared / delta_U_squared
    termB = ((l_U1U3**2 * l_P1P2**2 + l_U1U2**2 * l_P1P3**2) / (4 * area_3d) - (dot_U * dot_P)**2 / (2 * area_3d))
    distortion_metric_gradient = d_termA_d_U1 * termB + termA * d_termB_d_U1
    
    return conformal_distortion, authalic_distortion, term1, distortion_metric, distortion_metric_gradient

def objective_function(parameterized_vertices_2d_flat):

    parameterized_vertices_2d = parameterized_vertices_2d_flat.reshape((-1, 2))
    
    total_distortion_metric = 0
    total_distortion_metric_gradient = np.zeros_like(parameterized_vertices_2d)
    
    for face in faces:
        triangle_3d = vertices[np.array(face)]
        triangle_2d = parameterized_vertices_2d[np.array(face)]
        
        _, _, _, distortion_metric, distortion_metric_gradient = compute_distortion_metrics(triangle_3d, triangle_2d)
        
        total_distortion_metric += distortion_metric
        total_distortion_metric_gradient[np.array(face)] += distortion_metric_gradient
    
    
    return total_distortion_metric, total_distortion_metric_gradient.flatten()
```

```python
from scipy.optimize import minimize
import numpy as np

noise_amplitude = 1e-6
initial_params = vertices[:, :2] + noise_amplitude * np.random.randn(*vertices[:, :2].shape)


result = minimize(
    objective_function, 
    initial_params.flatten(),  
    method='trust-constr', 
    jac=True, 
    options={'maxiter': 10000} 
)


optimal_params = result.x.reshape((-1, 2))

print("Optimized parameters:", optimal_params)
print("Optimization success:", result.success)
print("Objective function value at optimum:", result.fun)
```

```python
import matplotlib.pyplot as plt

x = optimal_params[:, 0]
y = optimal_params[:, 1]

# Plot the vertices
plt.scatter(x, y, c='blue')

# Plot the edges connecting the vertices to form faces
for face in faces:
    for i in range(len(face)):
        for j in range(i+1, len(face)):
            plt.plot([x[face[i]], x[face[j]]], [y[face[i]], y[face[j]]], c='red')

plt.show()
```
